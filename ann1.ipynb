```python
# Assignment 1: Write a program to scheme a few activation functions that are used in neural networks
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#-10 is the start of the interval.
# 10 is the end of the interval.
# 1000 is the number of points to generate between the start and end points, inclusive.
x = np.linspace(-10, 10, 1000)
```

```python
def linear_function(x):
    return x

def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))

def tanh_function(x):
    return np.tanh(x)

def relu_function(x):
    return np.maximum(0, x)
```

```python
linear = linear_function(x)
sigmoid = sigmoid_function(x)
tanh = tanh_function(x)
relu = relu_function(x)
```

```python
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.plot(x, linear)
plt.title('Linear')

plt.subplot(2, 2, 2)
plt.plot(x, sigmoid)
plt.title('Sigmoid')

plt.subplot(2, 2, 3)
plt.plot(x, tanh)
plt.title('Tanh')

plt.subplot(2, 2, 4)
plt.plot(x, relu)
plt.title('ReLU')

plt.tight_layout()
plt.show()
```
